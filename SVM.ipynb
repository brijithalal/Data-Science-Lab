```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
X = df.iloc[:, [2,3]]
Y = df.iloc[:, 4]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC #linear
lin = SVC(kernel = 'linear', random_state = 0)
lin.fit(X_train, Y_train)
pred = lin.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,pred))
```

```python
from sklearn.svm import SVC  #rbf
rbf = SVC(kernel = 'rbf', random_state = 0)
rbf.fit(X_train, Y_train)
pred = rbf.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,pred))
```

```python
from sklearn.svm import SVC  #poly
poly = SVC(kernel = 'poly', degree=4)
poly.fit(X_train, Y_train)
pred = poly.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,pred))
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1],c=Y_test)  

w = lin.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-2.5, 2.5)
yy = a * xx - (poly.intercept_[0]) / w[1] #plot based on poly kernel

plt.plot(xx, yy)
plt.axis("off"), plt.show();

```
