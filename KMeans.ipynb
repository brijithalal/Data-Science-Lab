```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
mydata=pd.read_csv("Mall_Customers.csv")
x=mydata.iloc[:,[3,4]].values
mydata.head()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km.fit(x)
pred=km.predict(x)
pred
```

```python
plt.scatter(x[pred==0,0],x[pred==0,1],c='blue',label='cluster1')
plt.scatter(x[pred==1,0],x[pred==1,1],c='yellow',label='cluster2')
plt.scatter(x[pred==2,0],x[pred==2,1],c='red',label='cluster3')
plt.scatter(x[pred==3,0],x[pred==3,1],c='green',label='cluster4')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], c='purple' , label='Centroids')
plt.legend()
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1],c=pred) #or method
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], c='purple' , label='Centroids')
plt.show()
```
